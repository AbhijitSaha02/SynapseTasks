```python
# Importing OpenCV and cv2_imshow from Google colab since it crashes on using the cv2.imshow() function
import cv2
from google.colab.patches import cv2_imshow
 
# Reading the image from the drive
image= cv2.imread('/content/drive/MyDrive/Container.png')
original_image= image
 
# Converting into grayscale image
gray= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
 
cv2_imshow(gray)
```

```python
# Finding the edges using canny edge detection technique
edges= cv2.Canny(gray, 50,200)
 
cv2_imshow(edges)
```

```python
 
# Finding the contours in the image
contours, hierarchy= cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
print ('Number of contours found = ', len(contours))
 
# Defining a function which returns the list of all rectangles in the image
def get_contour_areas(contours):
 
    all_areas= []
 
    for cnt in contours:
        area= cv2.contourArea(cnt)
        all_areas.append(area)
 
    return all_areas
 
# Sorting the contours in ascending order
sorted_contours= sorted(contours, key=cv2.contourArea)
 
# Storing the largest rectangle in another variable
largest_item= sorted_contours[-1]
 
# Drawing contours on the largest rectangle
cv2.drawContours(original_image, largest_item, -1,(0,0,255),5)
cv2.waitKey(0)
print('\nLargest rectangle is:')
 
# Displaying the image
cv2_imshow(original_image)
 
 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
